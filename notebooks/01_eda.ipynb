```python
import sys, pathlib
ROOT = pathlib.Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.data.load_data import load_raw, basic_eda
from src.features.feature_engineering import add_engineered_features
from src.utils.paths import load_config

cfg = load_config()
df = load_raw()
df.head()
```

```python
eda = basic_eda(df)
print('資料維度：', eda['shape'])
print('故障比例：', eda['target_rate'])
print('產品類別分布：', eda['type_distribution'])
print('故障類型計數：', eda['failure_type_counts'])
```

```python
df_eng = add_engineered_features(df)
df_eng[['temp_diff','power_proxy','wear_torque_interaction',
        'wear_speed_interaction','temp_wear_interaction']].describe().round(3)
```

```python
import seaborn as sns, matplotlib.pyplot as plt
sns.set_theme(style='whitegrid')
fig, ax = plt.subplots(figsize=(5,4))
sns.countplot(data=df, x='Machine failure', ax=ax)
ax.set_title('Machine failure 分布')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,6))
corr = df_eng[cfg['columns']['numeric_raw'] + ['temp_diff','power_proxy',
        'wear_torque_interaction','wear_speed_interaction',
        'temp_wear_interaction']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, ax=ax)
plt.tight_layout(); plt.show()
```
